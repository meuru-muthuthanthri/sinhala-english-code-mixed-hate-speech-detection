```python
import pickle
import seaborn as sns
import matplotlib.pyplot as plt

# Defining the limits for the colormap
vmin = 0
vmax = 1
nrows = 7
ncols = 3
i = 0
j = 0

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

def increment_i_j():
    global i, j
    if j < (ncols - 1):
        j = j + 1
    else:
        j = 0
        i = i + 1

def plot_graph(name, file_name):
    # load the values from a file
    with open(file_name, 'rb') as f:
        data = pickle.load(f)
    cm_percentage = data['cm_percentage']
    
    ax = axs[i][j]
    sns.heatmap(cm_percentage, annot=True, fmt='.2%', cmap='Blues', ax=ax, cbar=False, vmin=vmin, vmax=vmax)
    increment_i_j()
    ax.set_title(name)
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')

collect_data('Logistic Regression', '2. Logistic Regression.pkl')
collect_data('KNN', '2. k-NN.pkl')
collect_data('SVM', '2. SVM.pkl', )
plot_graph('AdaBoost', '2. AdaBoost.pkl')
plot_graph('Multinomial Naive Bayes', '2. Multinomial Naive Bayes.pkl')
plot_graph('Decision Tree', '2. Decision Tree.pkl')
plot_graph('Perceptron', '2. Perceptron.pkl')
plot_graph('Random Forest Decision Tree', '2. Random Forest Decision Tree.pkl')
plot_graph('FastText with multinomial logistic regression', '2.1 Fasttext.pkl')
plot_graph('CNN with FastText existing model', '2.3 CNN with fasttext existing model.pkl')
plot_graph('CNN with FastText model training', '2.4 CNN with fasttext model training.pkl')
plot_graph('Bag of N-words with Linear Regression', '2.5 Bag of N-words with Linear Regression.pkl')
plot_graph('RNN with LSTM', '2.7 RNN with LSTM.pkl')
plot_graph('CNN with integer encoding', '2.6 CNN.pkl')
plot_graph('XGBoost', '2.8 XGBoost.pkl')
plot_graph('LightGBM', '2.9 LightGBM.pkl')
plot_graph('CatBoost', '2.10 CatBoost.pkl')
plot_graph('Bert', '2.11 Bert.pkl')
plot_graph('DistilBert', '2.12 DistilBert.pkl')
plot_graph('RoBERTa', '2.13 RoBERTa.pkl')
plot_graph('GPT2', '2.14 GPT2.pkl')

# axs[5][1].axis('off')
# axs[5][2].axis('off')
# axs[5][3].axis('off')
# Adding a common color bar
cbar = fig.colorbar(axs[0][0].collections[0], ax=axs, location='right', use_gridspec=False, pad=0.1)
cbar.set_ticklabels(['00%', '20%', '40%', '60%', '80%', '100%'])

plt.show()
```
